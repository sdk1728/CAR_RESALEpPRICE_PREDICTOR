```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('car data.csv')
df.head(3)
```

```python
df.info()
```

```python
df.describe()
```

```python
import datetime
date_year=datetime.datetime.now()
df['age']=date_year.year-df['Year']
```

```python
df.head(2)
```

```python
sns.boxplot(x='Selling_Price',data=df)
```

```python
sorted(df['Selling_Price'],reverse=True)
# need to drop values which are greater than 6000000 for more accurate result
```

```python
df=df[df['Selling_Price']<=25]
df.describe()
```

```python
df.head(3)
```

```python
drop_columns=['Car_Name','Year']
df=df.drop(columns=drop_columns,index=1)
```

```python
df.head(3)
```

```python
df.shape
```

```python
fuel_types=df['Fuel_Type'].unique()
seller_types=df['Seller_Type'].unique()
transmission_types=df['Transmission'].unique()
owner_types=df['Owner'].unique()
print('Fuel_types:',fuel_types)
print('Seller_types:',seller_types)
print('Transmission_types:',transmission_types)
print('Owner_types:',owner_types)
```

```python
df_encoded=pd.get_dummies(df,columns=['Fuel_Type','Seller_Type','Transmission'])
df_encoded.head(3)
```

```python
df_encoded.rename(columns={'Present_Price':'Buying_price'},inplace=True)
```

```python
df_encoded.head(3)
```

```python
df_encoded.info()
```

```python
x=df_encoded.drop(columns='Selling_Price',axis=1)
y=df_encoded['Selling_Price']
```

```python
x.head(3)
```

```python
y.head(3)
```

```python
#splitting the dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
%pip install xgboost
```

```python
#importing the models
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
# Model training
lr=LinearRegression()
lr.fit(x_train,y_train)

rf=RandomForestRegressor()
rf.fit(x_train,y_train)

gb=GradientBoostingRegressor()
gb.fit(x_train,y_train)

xgbr=XGBRegressor()
xgbr.fit(x_train,y_train)
```

```python
y_pred1=lr.predict(x_test)
y_pred2=rf.predict(x_test)
y_pred3=gb.predict(x_test)
y_pred4=xgbr.predict(x_test)
```

```python
# evaluating models
from sklearn import metrics
score1=metrics.r2_score(y_test,y_pred1)
score2=metrics.r2_score(y_test,y_pred2)
score3=metrics.r2_score(y_test,y_pred3)
score4=metrics.r2_score(y_test,y_pred4)
```

```python
print('Liner_reg:',score1,'\n','Random_for_reg:',score2,'\n','Grad_boost_reg:',score3,'\n','XGB_reg:',score4)
```

```python
#Saving final model
GB=GradientBoostingRegressor()
GB_final=GB.fit(x,y)
```

```python
import joblib
joblib.dump(GB_final,'Car_price_predictor')
```
